```python
import seaborn as sns#Seaborn kütüphanesini import ediyoruz.
df=sns.load_dataset('diamonds')#Diamonds isimli veri setini seaborn kütüphanesi içerisinden 
#çekiyoruz.
df=df.select_dtypes(include=['float64','int64'])#Sadece içerisindeki sayısal değişkenlere 
#erişmek için bir tip seçimi gerçekleştiriyoruz.
df=df.dropna()#Veri setinin içerisinde yer alan eksik değerleri direk silerek, dışarıda 
#bırakarak son dataframemimizi oluşturuyoruz.Burada aykırı gözlemleri,aykırı değerleri yakalama işlemini 
#rahat bir şekilde yapabilmek adına şimdilik direk eksik gözlemleri,değerleri silmiş olduk. 
df.head()#Veri setimizi gözlemliyoruz.
```

```python
import numpy as np#Numpyı import ediyoruz.
from sklearn.neighbors import LocalOutlierFactor#Local outlier factor fonksiyonumuzu da sklearn isimli 
#kütüphanenin içerisinden neighbors içerisinden import ediyoruz.
```

```python
clf = LocalOutlierFactor(n_neighbors = 20, contamination = 0.1)#Local outlier factor fonksiyonumuzu 
#kullanarak ilerde bir skorlama işlemi gerçekleştireceğimiz bir nesne oluşturuyoruz bunuda clf nesnesi
#olarak isimlendiriyoruz.LOF fonksiyonumuzun içerisine komşuluk sayısını ifade eden n_neighbors argümanını
#girip bunu 20 olarak belirliyoruz.Contamination isimli yoğunluğu ifade eden argümanı da girip, bunada 0.1
#değerini giriyoruz.Yani 20 komşu ve 0.1 yoğunluk değerleri ile LOF fonksiyonumuzu çalıştırıyoruz.
```

```python
clf#Clf nesnemizi gözlemliyoruz ve LOF algoritmamızın biçimsel özelliklerine erişiyoruz.
```

```python
clf.fit_predict(df)#LOF algoritmamızın biçimsel özelliklerini barındıran clf nesnemizi kullanarak
#ve fit_predict fonksiyonunun içerisine çok değişkenli yaklaştığımızdan dolayı dataframemimizi girerek
#bir fit etme işlemi yani algoritmayı çalıştırma işlemi gerçekleştiriyoruz.
```

```python
df_scores = clf.negative_outlier_factor_#clf nesnesine negative outlier factor fonksiyonunu uyguluyoruz.
#Böylece veri setimizdeki her bir gözlem biriminin Local outlier factor skorunu elde ediyoruz.Bunuda
#df_scores olarak kaydediyoruz.
```

```python
df_scores#df_scores'larımızı gözlemliyoruz.
```

```python
np.sort(df_scores)[0:20]#numpy arrayimizi sort fonksiyonuyla küçükten büyüğe sıraladıktan sonra ilk 20 
#değerimizi gözlemliyoruz.
```

```python
np.sort(df_scores)[13]#Eşik değer belirleme işlemlerine geçiyoruz.Örneğin arrayimizin küçükten büyüğe
#sıralanmış halinin 13.indexine karşılık gelen skoru gözlemliyoruz.
```

```python
esik_deger = np.sort(df_scores)[13]#Gözlemlemiş olduğumuz skoru yani değeri eşik değer olarak belirliyoruz
```

```python
df_scores < esik_deger#Aykırı değerlere erişme işlemlerine geçiyoruz.Eşik değerden küçük olan değerler
#aykırı değerlerimiz olmuş olacak ve eşik değerden küçük olan değerleri gözlemliyoruz.
```

```python
aykiri_tf = df_scores < esik_deger#Eşik değerden küçük değerleri gözlemleme işlemini aykiri_tf olarak 
#atıyoruz.
```

```python
aykirilar=df[aykiri_tf]#Aykiri tf'i dataframemimizin içerisine gönderip true olanları yakalayıp aykırı
#değerlere erişiyoruz.Bu işlemi aykirilar olarak adlandırıyoruz.
```

```python
aykirilar#Aykırı değerleri gözlemliyoruz.
```

# SİLME YÖNTEMİ

```python
df_scores > esik_deger#Aykırı olmayan değerlere erişme yani aykırı değerleri silme işlemi için eşik
#değerden büyük olan değerleri gözlemliyoruz. 
```

```python
yeni_df=df[df_scores > esik_deger]#Eşik değerden büyük olan değerlerin gerçek değerlerine yani aykırı 
#olmayan değerlere, dataframemimizin içerisine eşik değerden büyük olan değerlerin arraylerini girip 
#erişiyoruz.Bunu da yeni_df olarak atıyoruz.  
```

```python
yeni_df#yeni_df'imizi yani aykırı olmayan değerlerimizi gözlemliyoruz.
```

# BASKILAMA YÖNTEMİ

```python
df[df_scores == esik_deger]#Eşik değer skorumuzun karşılığı olan verisetimizdeki gözlem birimimize
#erişiyoruz.
```

```python
baski_deger = df[df_scores == esik_deger]#Bu gözlem birimimizi baski_değer olarak kaydediyoruz.
```

```python
aykirilar = df[aykiri_tf]#Aykırılar komutumuzu yazıyoruz.
```

```python
aykirilar#Aykırı değerlerimizi gözlemliyoruz. 
```

```python
#Aykırı değerlerimizi baskı değeri ile baskılama işlemini yapabilmek için aykırılar dataframemimizi ve
#baskı değer dataframemimizi indexsiz numpy arraylerine çevirmemiz gerekiyor.Bunun için;
aykirilar.to_records(index = False)#Aykırılar dataframemimizi to records fonksiyonu ile index 
#parametresini de false olarak girip indexlerini kaldırdığımız bir  numpy array'ine çeviriyoruz.
```

```python
res = aykirilar.to_records(index = False)#Aykırılar dataframemimizi array'e çevirme işlemini yani
#yukarıdaki işlemimizi res diye kaydediyoruz.
```

```python
res[:] = baski_deger.to_records(index = False)#Res'in içindeki tüm değerleri to records fonksiyonu ile 
#array'leştirdiğimiz ve index değerlerini kaldırdığımız baskı değer arrayiyle değiştiriyoruz. 
```

```python
res#Res array'imizi gözlemliyoruz.
```

```python
#Fakat yukarıdaki işlemimiz gerçek veri setimizdeki aykırı değerlerde bir değişiklik göstermez çünkü biz 
#bu işlemleri arrayler üzerinde yaptık.Bunun için aykırılar dataframemimizi yeniden tanımlamamız
#gerekiyor.
df[aykiri_tf]#Gerçek verisetimizi(ana dataframemimizi) gözlemliyoruz.
```

```python
#Şimdi aykırılar dataframemini yeniden tanımlama işlemine geçiyoruz.
import pandas as pd#Pandası import ediyoruz.
df[aykiri_tf] = pd.DataFrame(res, index = df[aykiri_tf].index)#Res array'ini ve Res array'inde index
#olmadığı için, belirlediğimiz index değerlerini girip aykırılar dataframemini yeniden tanımlıyoruz.
```
